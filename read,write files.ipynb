```python
import sys

def read_file(path):
    """Return the contents of a text file."""
    with open(path, 'r', encoding='utf-8') as f:
        return f.read()

if __name__ == "__main__":
    # 1. Filter out Jupyter's internal '--f=' argument
    args = [arg for arg in sys.argv if not arg.startswith('--f=')]
    
    # 2. Check if a filename was actually passed after filtering
    if len(args) >= 2:
        filename = args[1]
    else:
        # Fallback to your default path if no filename is provided
        filename = r"C:\Users\sowmy\OneDrive\Documents\sample.txt"

    try:
        content = read_file(filename)
        print(content)
    except Exception as e:
        print(f"Error reading {filename}: {e}")
```

```python
# Read the Example1.txt
example1 = r"C:\Users\sowmy\OneDrive\Documents\sample.txt"
file1 = open(example1, "r")
print(file1.read(4))
print(file1.read(4))
print(file1.read(7))
print(file1.read(15))
file1.close()
```

```python
with open(example1, 'w') as writefile:
    writefile.write("This is line A\n")
    writefile.write("This is line B\n")
```

```python
with open(example1, 'r+') as testwritefile:
    testwritefile.seek(0,0) #write at beginning of file
    testwritefile.write("Line 1" + "\n")
    testwritefile.write("Line 2" + "\n")
    testwritefile.write("Line 3" + "\n")
    testwritefile.write("Line 4" + "\n")
    testwritefile.write("finished\n")
    testwritefile.seek(0,0)
    print(testwritefile.read())
```

```python
with open(example1, 'a+') as testwritefile:
    testwritefile.write("This is line E\n")
    testwritefile.seek(0,0) #move cursor to beginning of file
    print(testwritefile.read())
```
